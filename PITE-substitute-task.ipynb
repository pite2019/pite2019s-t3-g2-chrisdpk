```python
import sklearn
import sklearn.datasets
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
boston = sklearn.datasets.load_boston()
df = pd.DataFrame(data = boston.data)
df.head()
```

```python
df.hist(column=5,bins=25)
df.hist(column=12, bins=25)
```

```python
lr = sklearn.linear_model.LinearRegression()
X = df.loc[:, [12,5]]
lr.fit(X, boston.target)
```

```python
plt.scatter(X[12], boston.target, color='b')
plt.plot(X, lr.predict(X), color='r')
plt.show()
plt.scatter(X[5], boston.target, color='orange')
plt.plot(X, lr.predict(X), color='purple')
plt.show()
```
